```python
import pandas as pd
import os

def load_and_extract_dataset(path, output_dir="extraction_output"):
    os.makedirs(output_dir, exist_ok=True)

    df = pd.read_csv(path)

    df.head().to_csv(
        f"{output_dir}/dataset_head.csv",
        index=False
    )

    return df

if __name__ == "__main__":
    load_and_extract_dataset("healthcare_dataset.csv")
```
