```python
import pandas as pd
import os
from collections import Counter

def text_analysis(df, text_columns, output_dir="text_analysis_output"):
    os.makedirs(output_dir, exist_ok=True)

    summary = []

    for col in text_columns:
        texts = df[col].dropna().astype(str)

        summary.append({
            "column": col,
            "total_records": texts.shape[0],
            "unique_values": texts.nunique(),
            "most_frequent_value": texts.value_counts().idxmax(),
            "most_frequent_count": texts.value_counts().max(),
            "avg_text_length": texts.str.len().mean()
        })

        word_list = []
        for text in texts:
            word_list.extend(text.lower().split())

        word_freq = Counter(word_list)
        pd.DataFrame(word_freq.most_common(50), columns=["word", "frequency"]) \
            .to_csv(f"{output_dir}/{col}_top_words.csv", index=False)

        texts.head(10).to_csv(
            f"{output_dir}/{col}_sample_texts.csv",
            index=False,
            header=[col]
        )

    pd.DataFrame(summary).to_csv(
        f"{output_dir}/text_analysis_summary.csv",
        index=False
    )

if __name__ == "__main__":
    df = pd.read_csv("healthcare_dataset.csv")

    text_columns = [
        "Medical Condition",
        "Medication",
        "Doctor",
        "Hospital",
        "Insurance Provider"
    ]

    text_analysis(df, text_columns)
```
