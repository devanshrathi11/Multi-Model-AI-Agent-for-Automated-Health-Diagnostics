```python
import pandas as pd
import os

df = pd.read_csv("healthcare_dataset.csv")
os.makedirs("parameter_output", exist_ok=True)

pd.DataFrame({"Available_Parameters": df.columns.tolist()}) \
  .to_csv("parameter_output/available_parameters.csv", index=False)

medical_params = ["Blood Type", "Medical Condition", "Medication", "Test Results"]
medical_data = []

for col in medical_params:
    for val in df[col].unique():
        medical_data.append({"Parameter": col, "Value": val})

pd.DataFrame(medical_data) \
  .to_csv("parameter_output/medical_parameters_unique.csv", index=False)

demo_params = ["Name", "Age", "Gender"]
df[demo_params].head() \
  .to_csv("parameter_output/demographics_sample.csv", index=False)

admin_params = [
    "Date of Admission",
    "Discharge Date",
    "Doctor",
    "Hospital",
    "Insurance Provider",
    "Billing Amount",
    "Room Number",
    "Admission Type"
]

df[admin_params].head() \
  .to_csv("parameter_output/administrative_sample.csv", index=False)

df.describe().transpose() \
  .to_csv("parameter_output/numerical_summary_statistics.csv")
```
