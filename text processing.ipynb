```python
import pandas as pd
import os
import re

def text_preprocess(df, text_columns, output_dir="text_process_output"):
    os.makedirs(output_dir, exist_ok=True)

    before = []
    for col in text_columns:
        for val in df[col].dropna().head(5):
            before.append({"column": col, "text": val})

    pd.DataFrame(before).to_csv(f"{output_dir}/before_text_processing.csv", index=False)

    for col in text_columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.lower()
            .str.replace(r"[^a-z0-9 ]", "", regex=True)
            .str.replace(r"\s+", " ", regex=True)
            .str.strip()
        )

    after = []
    for col in text_columns:
        for val in df[col].dropna().head(5):
            after.append({"column": col, "text": val})

    pd.DataFrame(after).to_csv(f"{output_dir}/after_text_processing.csv", index=False)

    df[text_columns].to_csv(f"{output_dir}/processed_text_columns.csv", index=False)

    return df

if __name__ == "__main__":
    df = pd.read_csv("healthcare_dataset.csv")

    text_cols = [
        "Medical Condition",
        "Medication",
        "Doctor",
        "Hospital",
        "Insurance Provider"
    ]

    text_preprocess(df, text_cols)
```
