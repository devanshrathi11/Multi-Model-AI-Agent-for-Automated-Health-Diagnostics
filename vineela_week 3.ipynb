```python
import pandas as pd

# ---------------------------
# 1. Load Week-2 Output
# ---------------------------
df = pd.read_csv("/content/sample_data/week2_cardio_pattern_analysis.csv", sep=',')

print("Week-2 Data Loaded Successfully")

# ---------------------------
# Generate Missing Columns for Week-3 Processing
# ---------------------------

# Convert age from days to years
df['age_years'] = (df['age'] / 365.25).astype(int)

# Calculate BMI and determine Obesity Risk
df['BMI'] = df['weight'] / ((df['height'] / 100)**2)
df['BMI_Status'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, float('inf')], labels=['Underweight', 'Normal', 'Overweight', 'Obese'], right=False)
df['Obesity_Risk'] = df['BMI'].apply(lambda x: "Yes" if x >= 30 else "No")

# Determine Blood Pressure Status and Hypertension Risk
df['Systolic_BP_Status'] = pd.cut(df['ap_hi'], bins=[0, 120, 129, 139, 180, float('inf')], labels=['Normal', 'Elevated', 'Hypertension Stage 1', 'Hypertension Stage 2', 'Hypertensive Crisis'], right=False)
df['Diastolic_BP_Status'] = pd.cut(df['ap_lo'], bins=[0, 80, 89, 120, float('inf')], labels=['Normal', 'Hypertension Stage 1', 'Hypertension Stage 2', 'Hypertensive Crisis'], right=False)
df['Hypertension_Risk'] = df.apply(lambda row: "Yes" if row['ap_hi'] >= 140 or row['ap_lo'] >= 90 else "No", axis=1)

# Determine Cholesterol/Glucose Status and Metabolic Risk
df['Cholesterol_Status'] = df['cholesterol'].map({1: 'Normal', 2: 'Above Normal', 3: 'Well Above Normal'})
df['Glucose_Status'] = df['gluc'].map({1: 'Normal', 2: 'Above Normal', 3: 'Well Above Normal'})
df['Metabolic_Risk'] = df.apply(lambda row: "Yes" if row['cholesterol'] > 1 or row['gluc'] > 1 else "No", axis=1)

# Determine Overall Risk Level based on 'cardio' column
df['Overall_Risk_Level'] = df['cardio'].map({1: 'High', 0: 'Low'})

# ---------------------------
# 2. Findings Synthesis Engine
# ---------------------------

def synthesize_findings(row):
    findings = []

    if row["Hypertension_Risk"] == "Yes":
        findings.append("Elevated blood pressure detected")

    if row["Obesity_Risk"] == "Yes":
        findings.append("Obesity based on BMI")

    if row["Metabolic_Risk"] == "Yes":
        findings.append("Metabolic risk due to high glucose and cholesterol")

    if row["Overall_Risk_Level"] == "High":
        findings.append("High cardiovascular risk")

    if not findings:
        return "All key health parameters are within acceptable ranges"

    return "; ".join(findings)

df["Health_Findings_Summary"] = df.apply(synthesize_findings, axis=1)

# ---------------------------
# 3. Personalized Recommendation Generator
# ---------------------------

def generate_recommendations(row):
    recommendations = []

    if row["Hypertension_Risk"] == "Yes":
        recommendations.append(
            "Reduce salt intake, manage stress, and monitor blood pressure regularly"
        )

    if row["Obesity_Risk"] == "Yes":
        recommendations.append(
            "Adopt a calorie-controlled diet and increase physical activity"
        )

    if row["Metabolic_Risk"] == "Yes":
        recommendations.append(
            "Limit sugar and saturated fat intake; follow a balanced diet"
        )

    if row["Overall_Risk_Level"] == "High":
        recommendations.append(
            "Consult a healthcare professional for further evaluation"
        )

    if not recommendations:
        return "Maintain current lifestyle with regular health check-ups"

    return "; ".join(recommendations)

df["Personalized_Recommendations"] = df.apply(generate_recommendations, axis=1)

# ---------------------------
# 4. Final Week-3 Output
# ---------------------------

final_columns = [
    "age_years",
    "Glucose_Status",
    "Cholesterol_Status",
    "Systolic_BP_Status",
    "Diastolic_BP_Status",
    "BMI_Status",
    "Overall_Risk_Level",
    "Health_Findings_Summary",
    "Personalized_Recommendations"
]

print("\nWEEK-3 SYNTHESIS & RECOMMENDATION OUTPUT\n")
print(df[final_columns].head())

# ---------------------------
# 5. Save Week-3 Output
# ---------------------------
df.to_csv("week3_health_synthesis_recommendations.csv", index=False)

print("\n✅ Week-3 Task Completed Successfully")
```
