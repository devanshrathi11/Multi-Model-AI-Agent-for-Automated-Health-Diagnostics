```python
import pandas as pd

# ==================================================
# 1. INPUT INTERFACE (Original Dataset)
# ==================================================
df = pd.read_csv("/content/sample_data/cardio_train.csv", sep=';')

# ==================================================
# 2. DATA PREPROCESSING
# ==================================================
df["age_years"] = (df["age"] / 365).astype(int)
df["BMI"] = df["weight"] / ((df["height"] / 100) ** 2)

# ==================================================
# 3. MODEL-1: PARAMETER INTERPRETATION (Week-1)
# ==================================================
df["Glucose_Status"] = df["gluc"].apply(lambda x: "Normal" if x == 1 else "High")
df["Cholesterol_Status"] = df["cholesterol"].apply(lambda x: "Normal" if x == 1 else "High")

df["Systolic_BP_Status"] = df["ap_hi"].apply(
    lambda x: "Normal" if 90 <= x <= 120 else "High"
)

df["Diastolic_BP_Status"] = df["ap_lo"].apply(
    lambda x: "Normal" if 60 <= x <= 80 else "High"
)

df["BMI_Status"] = df["BMI"].apply(
    lambda x: "Normal" if 18.5 <= x <= 24.9 else ("Overweight" if x < 30 else "Obese")
)

# ==================================================
# 4. MODEL-2: PATTERN & RISK ASSESSMENT (Week-2)
# ==================================================
df["Hypertension_Risk"] = df.apply(
    lambda r: "Yes" if r["Systolic_BP_Status"] == "High" or r["Diastolic_BP_Status"] == "High" else "No",
    axis=1
)

df["Obesity_Risk"] = df["BMI"].apply(lambda x: "Yes" if x >= 30 else "No")

df["Metabolic_Risk"] = df.apply(
    lambda r: "Yes" if r["Glucose_Status"] == "High" and r["Cholesterol_Status"] == "High" else "No",
    axis=1
)

def risk_score(row):
    score = 0
    if row["Hypertension_Risk"] == "Yes": score += 2
    if row["Obesity_Risk"] == "Yes": score += 2
    if row["Metabolic_Risk"] == "Yes": score += 2
    if row["age_years"] > 50: score += 1
    if row["smoke"] == 1: score += 1
    return score

df["Cardio_Risk_Score"] = df.apply(risk_score, axis=1)

df["Overall_Risk_Level"] = df["Cardio_Risk_Score"].apply(
    lambda x: "Low" if x <= 2 else ("Medium" if x <= 5 else "High")
)

# ==================================================
# 5. MODEL-3: FINDINGS SYNTHESIS (Week-3)
# ==================================================
def synthesize_findings(row):
    findings = []
    if row["Hypertension_Risk"] == "Yes":
        findings.append("High blood pressure")
    if row["Obesity_Risk"] == "Yes":
        findings.append("Obesity detected")
    if row["Metabolic_Risk"] == "Yes":
        findings.append("Metabolic risk present")
    if row["Overall_Risk_Level"] == "High":
        findings.append("High cardiovascular risk")

    return " | ".join(findings) if findings else "No major health risks detected"

df["Health_Summary"] = df.apply(synthesize_findings, axis=1)

# ==================================================
# 6. PERSONALIZED RECOMMENDATIONS (Week-3)
# ==================================================
def generate_recommendations(row):
    rec = []
    if row["Hypertension_Risk"] == "Yes":
        rec.append("Reduce salt intake and monitor blood pressure")
    if row["Obesity_Risk"] == "Yes":
        rec.append("Increase physical activity and manage weight")
    if row["Metabolic_Risk"] == "Yes":
        rec.append("Follow a balanced low-sugar, low-fat diet")
    if row["Overall_Risk_Level"] == "High":
        rec.append("Consult a healthcare professional")

    return " | ".join(rec) if rec else "Maintain healthy lifestyle"

df["Recommendations"] = df.apply(generate_recommendations, axis=1)

# ==================================================
# 7. FINAL REPORT GENERATION MODULE
# ==================================================
df["Medical_Disclaimer"] = (
    "This AI-generated report is for informational purposes only "
    "and is not a substitute for professional medical advice."
)

final_report = df[
    [
        "age_years",
        "Glucose_Status",
        "Cholesterol_Status",
        "Systolic_BP_Status",
        "Diastolic_BP_Status",
        "BMI_Status",
        "Overall_Risk_Level",
        "Health_Summary",
        "Recommendations",
        "Medical_Disclaimer"
    ]
]

# ==================================================
# 8. SAVE FINAL OUTPUT
# ==================================================
final_report.to_csv("week4_final_health_report.csv", index=False)

print("\n✅ WEEK-4 COMPLETED: Final Health Report Generated Successfully")
```
