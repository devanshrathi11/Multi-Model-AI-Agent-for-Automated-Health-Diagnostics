```python
import pandas as pd

# ---------------------------
# 1. Load Dataset
# ---------------------------
df = pd.read_csv("/content/sample_data/cardio_train.csv", sep=';')

print("Dataset Loaded Successfully")
print(df.head())

# ---------------------------
# 2. Data Preprocessing
# ---------------------------

# Convert age from days to years
df["age_years"] = (df["age"] / 365).astype(int)

# Calculate BMI
df["BMI"] = df["weight"] / ((df["height"] / 100) ** 2)

# ---------------------------
# 3. Reference Ranges (Model-1)
# ---------------------------

REFERENCE_RANGES = {
    "gluc": (1, 1),          # 1 = normal, 2/3 = high
    "cholesterol": (1, 1),  # 1 = normal, 2/3 = high
    "ap_hi": (90, 120),     # systolic BP
    "ap_lo": (60, 80),      # diastolic BP
    "BMI": (18.5, 24.9)
}

# ---------------------------
# 4. Parameter Interpretation Functions
# ---------------------------

def interpret_range(value, low, high):
    if pd.isna(value):
        return "Invalid"
    elif value < low:
        return "Low"
    elif value > high:
        return "High"
    else:
        return "Normal"

def interpret_glucose(value):
    if value == 1:
        return "Normal"
    elif value > 1:
        return "High"
    else:
        return "Invalid"

def interpret_cholesterol(value):
    if value == 1:
        return "Normal"
    elif value > 1:
        return "High"
    else:
        return "Invalid"

# ---------------------------
# 5. Apply Model-1 Interpretation
# ---------------------------

df["Glucose_Status"] = df["gluc"].apply(interpret_glucose)
df["Cholesterol_Status"] = df["cholesterol"].apply(interpret_cholesterol)

df["Systolic_BP_Status"] = df["ap_hi"].apply(
    lambda x: interpret_range(x, 90, 120)
)

df["Diastolic_BP_Status"] = df["ap_lo"].apply(
    lambda x: interpret_range(x, 60, 80)
)

df["BMI_Status"] = df["BMI"].apply(
    lambda x: interpret_range(x, 18.5, 24.9)
)

# ---------------------------
# 6. Final Output (Week-1 Result)
# ---------------------------

final_columns = [
    "age_years",
    "gender",
    "Glucose_Status",
    "Cholesterol_Status",
    "Systolic_BP_Status",
    "Diastolic_BP_Status",
    "BMI_Status"
]

print("\nWeek-1 Parameter Interpretation Output:\n")
print(df[final_columns].head())

# ---------------------------
# 7. Save Output
# ---------------------------

df.to_csv("week1_cardio_analysis_output.csv", index=False)

print("\n✅ Week-1 Task Completed Successfully")
```
