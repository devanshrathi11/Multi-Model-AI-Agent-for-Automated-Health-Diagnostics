```
import pdfplumber
import pytesseract
from pdf2image import convert_from_path

def extract_text_from_pdf(pdf_path):
    extracted = ""

    try:
        with pdfplumber.open(pdf_path) as pdf:
            print(f"\n Reading (text mode): {pdf_path}")
            for page_num, page in enumerate(pdf.pages, start=1):
                text = page.extract_text()

                if text and text.strip():
                    extracted += f"\n--- Page {page_num} (text) ---\n"
                    extracted += text + "\n"
                else:
                    print(f"   Page {page_num}: no text, using OCR...")
                    img = page.to_image(resolution=300).original
                    ocr_text = pytesseract.image_to_string(img)
                    extracted += f"\n--- Page {page_num} (ocr) ---\n"
                    extracted += ocr_text + "\n"

    except Exception as e:
        print(f"pdfplumber failed on {pdf_path}, using pure OCR. Error: {e}")
        pages = convert_from_path(pdf_path)
        for i, p in enumerate(pages, start=1):
            ocr_text = pytesseract.image_to_string(p)
            extracted += f"\n--- Page {i} (ocr only) ---\n"
            extracted += ocr_text + "\n"

    return extracted
```

```
import os

pdf_files = [f for f in os.listdir() if f.lower().endswith(".pdf")]
print("Found PDF files:", pdf_files)

all_text = ""

for pdf in pdf_files:
    print(f"Processing: {pdf}")
    pdf_text = extract_text_from_pdf(pdf)
    all_text += f"\n\n\n===== FILE: {pdf} =====\n\n"
    all_text += pdf_text

print("\n Finished extracting ")
print("Total characters extracted:", len(all_text))
```

```
import re

parameters = {
    "hemoglobin": r"hemoglobin[^0-9]*([0-9.]+)",
    "rbc": r"rbc[^0-9]*([0-9.]+)",
    "wbc": r"wbc[^0-9]*([0-9,]+)",
    "platelets": r"platelet[^0-9]*([0-9,]+)",
    "glucose": r"glucose[^0-9]*([0-9.]+)",
    "cholesterol": r"cholesterol[^0-9]*([0-9.]+)"
}

extracted_values = {}

for param, pattern in parameters.items():
    matches = re.findall(pattern, cleaned_text)
    extracted_values[param] = matches[:10]  # show first 10 found

extracted_values
```

```
import pandas as pd

df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in extracted_values.items()]))
df
```

```
import pandas as pd

df = pd.read_csv("extracted_parameters.csv")
print("Original data:")
print(df.head())

for col in df.columns:
    df[col] = df[col].astype(str).str.replace(",", "").str.strip()
    df[col] = pd.to_numeric(df[col], errors="coerce")

print("\nAfter basic cleaning (numeric conversion):")
print(df.head())

valid_ranges = {
    "hemoglobin": (5, 20),          # g/dL
    "rbc":        (2, 8),           # million cells/mcL
    "wbc":        (2000, 30000),    # cells/mcL
    "platelets":  (50000, 1000000), # platelets/mcL
    "glucose":    (40, 400),        # mg/dL
    "cholesterol":(50, 400)         # mg/dL
}

def classify_value(value, low, high):
    """Return a simple label for the value based on the range."""
    if pd.isna(value):
        return "missing"
    if value < low:
        return "low"
    if value > high:
        return "high"
    return "normal"

for col, (low, high) in valid_ranges.items():
    if col in df.columns:
        status_col = col + "_status"
        df[status_col] = df[col].apply(lambda v: classify_value(v, low, high))

print("\nWith status columns added:")
print(df.head())

main_params = [c for c in valid_ranges.keys() if c in df.columns]
df["non_missing_params"] = df[main_params].notna().sum(axis=1)
df_clean = df[df["non_missing_params"] > 0].drop(columns=["non_missing_params"])

print("\nCleaned and filtered data:")
print(df_clean.head())
df_clean.to_csv("validated_standardized_parameters.csv", index=False)
print("\nSaved validated_standardized_parameters.csv")
files.download("validated_standardized_parameters.csv")

```

```
import pandas as pd

df = pd.read_csv("validated_standardized_parameters.csv")
df.head()
```

```
def identify_patterns(row):
    insights = []

    # Anemia pattern
    if row.get("hemoglobin_status") == "low" and row.get("rbc_status") == "low":
        insights.append("possible_anemia_pattern")

    # High blood sugar pattern
    if row.get("glucose_status") == "high":
        insights.append("elevated_glucose_pattern")

    # Cholesterol risk pattern
    if row.get("cholesterol_status") == "high":
        insights.append("elevated_cholesterol_pattern")

    # Infection / inflammation pattern
    if row.get("wbc_status") == "high":
        insights.append("possible_infection_pattern")

    if not insights:
        insights.append("no_significant_pattern")

    return ", ".join(insights)
```

```
df["identified_patterns"] = df.apply(identify_patterns, axis=1)
df[["identified_patterns"]].head()
```

```
df.to_csv("milestone2_pattern_analysis.csv", index=False)
print("Saved milestone2_pattern_analysis.csv")
files.download("milestone2_pattern_analysis.csv")
```
